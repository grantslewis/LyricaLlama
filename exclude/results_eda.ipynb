```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

df = pd.read_csv('./selection_results.csv')
df.head()
```

```python
counts_df = df.copy()
counts_df['Favored_Base'] = counts_df[['0', '1', '2']].sum(axis=1)
counts_df['Favored_Lyrical'] = counts_df[['3', '4', '5']].sum(axis=1)
counts_df['Best'] = counts_df[['0','1','2','3','4','5']].apply(lambda x: x.argmax(), axis=1)
counts_df['Worst'] = counts_df[['0','1','2','3','4','5']].apply(lambda x: x.argmin(), axis=1)
counts_df.head()
```

```python
sel = counts_df[['Favored_Base', 'Favored_Lyrical', 'Song_title']] # , 'Best', 'Worst'
sel.set_index('Song_title', inplace=True)
sel.head()
```

```python
#sel.set_index("Song_title", inplace=True)

# Plot
sel.plot(kind='bar', figsize=(12, 6))
plt.title('Votes for Text Results by Song Title')
plt.xlabel('Song Title')
plt.ylabel('Number of Votes')
plt.xticks(rotation=45)
plt.show()

# Stacked Bar Plot
sel.plot(kind='bar', stacked=True, figsize=(12, 6))
plt.title('Stacked Votes for Text Results by Song Title')
plt.xlabel('Song Title')
plt.ylabel('Number of Votes')
plt.xticks(rotation=45)
plt.show()
```
